**import the libraries**

```python
import numpy as np
import pandas as pd
from google.colab.patches import cv2_imshow
import cv2
```

**Loading the image**

```python
image=cv2.imread("Cat.jpg",1)
```

```python
cv2_imshow(image)
```

**Shape of an image**

```python
image.shape
```

 **Converting an original image into gray-scale image**

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(image_gray)
```

**Inverting the image**

```python
image_invert=cv2.bitwise_not(image_gray)
```

```python
cv2_imshow(image_invert)
```

**Blurring the image using Guassian blur**

```python
image_blur=cv2.GaussianBlur(image_invert,(21,21),sigmaX=0,sigmaY=0)
```

```python
cv2_imshow(image_blur)
```

**Dividing the greyscale value of image by inverse of blurred image value**

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

**Final Image**

```python
final_image = dodgeV2(image_gray, image_blur)
```

```python
cv2_imshow(final_image)
```
